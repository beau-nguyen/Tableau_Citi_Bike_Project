```python
#import dependecies
import pandas as pd
```

```python
csv1 = "resources/citi_10_2019.csv"
csv2 = "resources/citi_09_2019.csv"
csv3 = "resources/citi_08_2019.csv"
```

```python
df1 = pd.read_csv(csv1)
df2 = pd.read_csv(csv2)
df3 = pd.read_csv(csv3)
```
